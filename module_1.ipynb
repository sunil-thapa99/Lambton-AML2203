```python
def split_df_datatype(df):
    object_df = df.select_dtypes('object')
    num_df = df.select_dtypes(['int64', 'float64'])

    return object_df, num_df
```

```python
object_df, num_df = split_df_datatype(data)
```

```python
object_df.info()
```

```python
num_df.info()
```

```python
df = data.copy()
```

```python
df = df.astype({'cyl': str})
```
